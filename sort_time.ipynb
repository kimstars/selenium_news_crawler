## Viết hàm sắp xếp tin tức theo thời gian

```python
from datetime import datetime
 
import re 


def convert_date_str_to_seconds(date_str):
    # Sử dụng regex để trích xuất thông tin về ngày, tháng và năm
    match = re.match(r'(\d{1,2}) thg (\d{1,2}), (\d{4})', date_str)
    
    if match:
        day, month, year = map(int, match.groups())
        date_object = datetime(year, month, day)
        current_time = datetime.now()
        time_difference = current_time - date_object
        return int(time_difference.total_seconds())
    
    return 0

def convert_relative_time_to_seconds(time_str):
    if 'ngày' in time_str:
        return int(time_str.split()[0]) * 24 * 60 * 60
    elif 'giờ' in time_str:
        return int(time_str.split()[0]) * 60 * 60
    elif 'tuần' in time_str:
        return int(time_str.split()[0]) * 7 * 24 * 60 * 60
    elif 'tháng' in time_str:
        return int(time_str.split()[0]) * 30 * 24 * 60 * 60
    else:
        return 0

def key_function(item):
    return convert_date_str_to_seconds(item['time']) if 'thg' in item['time'] else convert_relative_time_to_seconds(item['time'])




test_data = [{'name': 'Ngày mai 16-1, xét xử 100 bị cáo tấn công trụ sở UBND 2 xã ở Đắk Lắk',
  'time': '4 ngày trước',
  'link': 'https://tuoitre.vn/ngay-mai-16-1-xet-xu-100-bi-cao-tan-cong-tru-so-ubnd-2-xa-o-dak-lak-20240115113427307.htm',
  'content': None,
  'description': 'Ngày mai (16-1), TAND tỉnh Đắk Lắk sẽ đưa ra xét xử lưu động 100 bị cáo \ntrong vụ tấn công trụ sở UBND 2 xã ở Đắk Lắk rạng sáng 11-6-2023.'},
 {'name': 'Thấy gì từ phiên tòa xét xử 100 người vụ tấn công trụ sở xã ở Đắk Lắk?',
  'time': '2 ngày trước',
  'link': 'https://www.bbc.com/vietnamese/articles/cqv1n7287p3o',
  'content': None,
  'description': 'Phiên tòa xét xử vụ án nổ súng tấn công trụ sở chính quyền xã tại tỉnh Đắk \nLắk một lần nữa khơi lại những mâu thuẫn tích tụ lâu nay ở Tây...'},
 {'name': 'Vụ tấn công trụ sở UBND xã ở Đắk Lắk: Các luật sư nhất trí với bản luận tội',
  'time': '18 giờ trước',
  'link': 'https://laodong.vn/phap-luat/vu-tan-cong-tru-so-ubnd-xa-o-dak-lak-cac-luat-su-nhat-tri-voi-ban-luan-toi-1294355.ldo',
  'content': None,
  'description': 'Đắk Lắk - Liên quan đến vụ nổ súng tấn công trụ sở hai UBND xã Ea Tiêu và \nEa Ktur (huyện Cư Kuin) hôm 11.6.2023, tại phiên tòa,...'},
 {'name': 'Xét xử vụ khủng bố ở Đắk Lắk: Các bị cáo khai bị dụ dỗ, lôi kéo, hứa hẹn \ncho tiền',
  'time': '2 ngày trước',
  'link': 'https://congan.com.vn/vu-an/xet-xu-vu-khung-bo-o-dak-lak-cac-bi-cao-khai-bi-du-do-loi-keo-hua-hen-cho-tien_158024.html',
  'content': None,
  'description': '(CAO) Quá trình thẩm vấn công khai tại phiên toà, các bị cáo khai báo diễn \nbiến hành vi phạm tội, âm mưu thành lập cái gọi là "Nhà nước...'}]


# Sắp xếp test_data theo giá trị của key 'time'
sorted_data = sorted(test_data, key=key_function)

# In mảng sau khi sắp xếp
for data in sorted_data:
    print(data)


```
