```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

def preprocess_text(text):
    # Chuyển về chữ thường
    text = text.lower()

    # Loại bỏ dấu câu
    text = text.translate(str.maketrans("", "", string.punctuation))

    # Tokenization
    words = word_tokenize(text)

    # Loại bỏ stopwords
    stop_words = set(stopwords.words("english"))
    words = [word for word in words if word.lower() not in stop_words]

    return words

def calculate_similarity(input_text, headlines):
    input_words = preprocess_text(input_text)
    headline_words = [preprocess_text(headline) for headline in headlines]

    # Tính tần số xuất hiện của từng từ trong input text và mỗi headline
    input_word_freq = {word: input_words.count(word) for word in set(input_words)}
    headline_word_freqs = [{word: headline_words[i].count(word) for word in set(headline_words[i])} for i in range(len(headlines))]

    # Tính độ tương đồng bằng cách so sánh tần số xuất hiện của từng từ
    similarities = [sum(min(input_word_freq.get(word, 0), headline_word_freqs[i].get(word, 0)) for word in set(input_words)) for i in range(len(headlines))]

    return similarities

# Example usage
input_text = "vụ tấn công ở đắk lắk "
headlines = ["Vụ khủng bố tại tỉnh Đắk Lắk: 10 bị cáo nhận án chung thân", 
             "Xét xử nhóm khủng bố ở Đắk Lắk: Nhiều bị cáo khai bị dọa giết cả nhà nếu không tham gia tổ chức", 
             "Xét xử vụ khủng bố tại Đắk Lắk: Các bị cáo thừa nhận phạm tội do bị rủ rê, lôi kéo, kích động",
             "Bài viết về công nghệ và ứng dụng di động",
             "Xét xử lưu động 100 bị cáo vụ tấn công trụ sở UBND 2 xã ở Đắk Lắk, bắt đầu vào ngày mai 16-1",
             "Tòa tuyên án 100 bị cáo khủng bố tại Đắk Lắk",
             "Truy nã đặc biệt 6 đối tượng ở Đắk Lắk về tội khủng bố",
             "Thêm ba người tấn công trụ sở xã ở Đăk Lăk bị bắt"
             
             ]

similarities = calculate_similarity(input_text, headlines)

# Print the similarity scores
for i, similarity in enumerate(similarities):
    print(f"Similarity with headline {i + 1}: {similarity}")
```
